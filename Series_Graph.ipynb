```python
import math
import numpy as np
from matplotlib import pyplot as plt

# Sets the size of the window to draw in
plt.figure(figsize=(8,8))

#The following code sets up the axes system to draw on
tMin = -5.0
tMax = 5.0
fMin = -10
fMax = 10
plt.axis([tMin, tMax, fMin, fMax])
plt.ylabel('f', fontsize=20)
plt.xlabel('t', fontsize=20)
plt.title('Series VS Function Graph', fontsize=20)

def f(t):
  #return(math.exp(t))  #e^t
  #if(-1.0 < t):
    #return(math.log(1.0+t)) #ln(1+t)
  if(t != -1.0):
    return(1/(1-t))

def f0(t):
  order = 0
  sum = 0.0
  n=0
  while n <= order:
    #sum = sum + (t**n)/math.factorial(int(n))  #e^t
    #sum = sum + ((-1)**n)*(t**(n+1))/(n+1)  #ln(1+t)
    sum = sum + (t**n)  #1/(1-t)
    n = n+1
  return(sum)

def f1(t):
  order = 1
  sum = 0.0
  n=0
  while n <= order:
    #sum = sum + (t**n)/math.factorial(int(n))  #e^t
    #sum = sum + ((-1)**n)*(t**(n+1))/(n+1)  #ln(1+t)
    sum = sum + (t**n)  #1/(1-t)
    n = n+1
  return(sum)

def f2(t):
  order = 2
  sum = 0.0
  n=0
  while n <= order:
    #sum = sum + (t**n)/math.factorial(int(n))  #e^t
    #sum = sum + ((-1)**n)*(t**(n+1))/(n+1)  #ln(1+t)
    sum = sum + (t**n)  #1/(1-t)
    n = n+1
  return(sum)

def f3(t):
  order = 3
  sum = 0.0
  n=0
  while n <= order:
    #sum = sum + (t**n)/math.factorial(int(n))  #e^t
    #sum = sum + ((-1)**n)*(t**(n+1))/(n+1)  #ln(1+t)
    sum = sum + (t**n)  #1/(1-t)
    n = n+1
  return(sum)

def f4(t):
  order = 4
  sum = 0.0
  n=0
  while n <= order:
    #sum = sum + (t**n)/math.factorial(int(n))  #e^t
    #sum = sum + ((-1)**n)*(t**(n+1))/(n+1)  #ln(1+t)
    sum = sum + (t**n)  #1/(1-t)
    n = n+1
  return(sum)

def fAlot(t):
  order = 25
  sum = 0.0
  n=0
  while n <= order:
    #sum = sum + (t**n)/math.factorial(int(n))  #e^t
    #sum = sum + ((-1)**n)*(t**(n+1))/(n+1)  #ln(1+t)
    sum = sum + (t**n)  #1/(1-t)
    n = n+1
  return(sum)

t = tMin
dt = 0.01
while t < tMax:
  lines = plt.plot([t, t+dt], [f(t),f(t+dt)], color='k', linewidth=1.0)
  lines = plt.plot([t, t+dt], [f0(t),f0(t+dt)], color='r', linewidth=1.0)
  lines = plt.plot([t, t+dt], [f1(t),f1(t+dt)], color='b', linewidth=1.0)
  lines = plt.plot([t, t+dt], [f2(t),f2(t+dt)], color='g', linewidth=1.0)
  lines = plt.plot([t, t+dt], [f3(t),f3(t+dt)], color='m', linewidth=1.0)
  lines = plt.plot([t, t+dt], [f4(t),f4(t+dt)], color='c', linewidth=1.0)
  lines = plt.plot([t, t+dt], [fAlot(t),fAlot(t+dt)], color='y', linewidth=1.0)
  t = t + dt

plt.show()
        

```
