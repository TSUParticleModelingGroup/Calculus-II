```python
# This code runs Euler's method. Which can move a differintial equation (DE) through time even if you can not solve the DE.

# G is the DE which takes in a time and function value and returns a change in function value
# In other words this DE is f'=f+t. Our book usually uses G for this so I did here as well to help you connect with your book.
def G(f,t):
    return(f+t)

# These are your initial conditions.
fStart = 1.0  # This is your starting function value. Is physics terms this is your initial position. 
tStart = 0.0  # This is your starting time.
print("Initial", tStart,fStart) # Printing out the initial conditions.

tStop = 1.0  # This is your stop time
dt = 0.1  # This is your jump size.

f = fStart
t = tStart
while t < tStop:
    df = G(f,t)  # Using the DE to see how fast the function is changing.
    fNew = f + dt*df  # Your new function value is the old value plus how much it changed (Rate of change X time it was changing).
    tNew = t + dt  # Your new time is the old time plus the change in time.
    print(tNew,fNew)  # Printing the new values.
    t = tNew  
    f = fNew
print(tNew,fNew)
```
