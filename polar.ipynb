```python
import math
import numpy as np
from matplotlib import pyplot as plt

# Sets the size of the window to draw in
plt.figure(figsize=(8,8))

#The following code sets up the axes system to draw on
xMin = -3.0
xMax = 3.0
yMin = -3.0
yMax = 3.0
plt.axis([xMin, xMax, yMin, yMax])
plt.ylabel('y', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.title('Parametric Graphs', fontsize=20)

# This is your angle range I left it as t because python does not take greek letters.
tMin = -1.0
tMax = 20.0

# This is the polar function you will be graphing
def r(t):
  return(math.cos(2.0*t)*math.sin(12*t))
  #return(math.sqrt(t))
  #return(t)
  #return(t**3-3.0*t)

# Taking from polar to rectangular to use python's lines function.
def x(t):
  return(r(t)*math.cos(t))

def y(t):
  return(r(t)*math.sin(t))

t = tMin
dt = 0.01
while t < tMax:
  lines = plt.plot([x(t), x(t+dt)], [y(t),y(t+dt)], color='k', linewidth=1.0)
  t = t + dt

plt.show()
        

```
