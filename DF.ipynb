```python
import math
import numpy as np
from matplotlib import pyplot as plt

# opens the graphic in a new window
#%matplotlib qt5

# Sets the size of the window
plt.figure(figsize=(8,8))

# Setting up the axes we will be drawing on
tStart = -5
tStop = 5
fStart = -5
fStop = 5
plt.axis([tStart, tStop, fStart, fStop])
plt.ylabel('f', fontsize=20)
plt.xlabel('t', fontsize=20)
plt.title('Direction Fields', fontsize=20)

def differentialEquation(f,t):
    "this is the differential equation that will generate the DF"
    #return(0.2*f)
    #return((5-f)*f)
    #return((.2/5)*(5-f)*f)
    #return(f+t)
    #return(f)
    #if t==1:
        #t = 1.0001
    #return(t/(f*f-4))
    #if f==0:
        #f = 0.0001
    #return(t/f)
    #return((f-1)*2*t)
    #return((np.cbrt(f+2))*2*t)
    return(f/(t*t-9))

gridt = 0.3
gridf = 0.3
dt = gridt

t = tStart
while t <= tStop:
    f = fStart
    while f <= fStop:
        plt.plot([t], [f], 'ko', ms = 2.0)
        df = differentialEquation(f,t)*dt
        length = math.sqrt(df*df + dt*dt)/dt #makes the arrows have a length of the t grid
        plt.arrow(t, f, 0.7*(dt/length), 0.7*(df/length), head_width=0.04, head_length=0.08, fc='g', ec='purple')
        f += gridf
    t += gridt

plt.show()
```
