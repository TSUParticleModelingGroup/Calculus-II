```python
import math
import numpy as np
from matplotlib import pyplot as plt

# Sets the size of the window to draw in
plt.figure(figsize=(8,8))

#The following code sets up the axes system to draw on
xMin = -10.0
xMax = 10.0
yMin = -10.0
yMax = 10.0
plt.axis([xMin, xMax, yMin, yMax])
plt.ylabel('y', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.title('Parametric Graphs', fontsize=20)

tMin = -1.0
tMax = 10.0

def x(t):
  #return(math.exp(t))
  return(4.0*math.cos(3.0*t))
  #return(math.sqrt(t))
  #return(t)
  #return(t**3-3.0*t)

def y(t):
  #return(math.exp(t))
  return(5.0*math.sin(3.0*t))
  #return(t**3+2.0*t)
  #return(t**2)
  #return(t**2 -3.0)

t = tMin
dt = 0.01
while t < tMax:
  lines = plt.plot([x(t), x(t+dt)], [y(t),y(t+dt)], color='k', linewidth=1.0)
  t = t + dt

plt.show()
        

```
